```python
import math
lis=[i for i in range(1,21)]
print("List of the first 20 natural numbers:- "+str(lis))

even=sum(i for i in range(0,21,2))
print("Sum of all even numbers in the list:- ", even)

odd=[i for i in range(1,21,2)]
pro_odd=1
for r in odd:
    pro_odd*=r
print("Product of all odd numbers in the list:- ", pro_odd)

mean=sum(lis)/len(lis)
median=(lis[len(lis)//2]+lis[(len(lis)//2)-1])/2
squared_diffs = [(x - mean) ** 2 for x in lis]
variance = sum(squared_diffs)/len(lis)
std = math.sqrt(variance)
print("Mean:- ", mean)
print("Median:- ",median)
print("vairence:- ",variance)
print("Standard Deviation:- ",std)
```

```python
n=int(input("Input:- "))
def Fibonacci(n):
    y=[0]
    i=0
    s=1
    for i in range(n-1):
      y.append(s)
      t=s
      s+=i
      i=t
    return y
x=[i for i in range(1,n+1)]
y=Fibonacci(n)
print("First "+str(n)+" Fibonacci numbers:- "+str(y))
import matplotlib.pyplot as plt
fig, ax = plt.subplots()
ax.plot(x, y)
plt.show()
```

```python
def cp(string):                                               #Q6
    string=string.lower()
    lis=list(string)
    while True:
        try:
            lis.remove(" ")
        except:
            break
    t=len(lis)//2
    for i in range(t):
        if lis[i]!=lis[-(i+1)]:
            return False
        t-=1
    return True
print("Output: "+str(cp(input("Input:- "))))
```

```python
def merge(d1,d2):                                     #Q9
    for i in d2:
        try:
            if d1[i]:
                d1[i]+=d2[i]
        except:
            d1[i]=d2[i]
    return d1
d1 = {'a': 10, 'b': 20}
d2 = {'b': 5, 'c': 15}
print("Output: "+str(merge(d1, d2)))
```

```python
def find_anagrams(word, words):
    word_list=list(word)
    word_list=sorted(word_list)
    result=[]
    for i in words:
        if word_list==sorted(list(i)):
            result.append(i)
    return result


word, words = "listen", ["enlist", "google", "inlets", "banana"]

print("Output: ", find_anagrams(word, words))
```

```python
a=['hello','hi','hola','water']
b=['water','oil','liquid','paint']
A=set(a)
B=set(b)
intersection=list(A&B)
union=list(A|B)
symmetric_difference=list(A|B - A&B)
print("Intersection:- ",intersection)
print("Union:- ", union)
print("symmetric difference:- ", symmetric_difference)

from matplotlib_venn import venn2
from matplotlib import pyplot as plt

ven=venn2([A,B], ("List 1", "List 2"))

for idx, subset in enumerate(['10', '01', '11']):      # for showing text of each word in venn diagram
    elements = None
    if subset == '10':
        elements = A - B
    elif subset == '01':
        elements = B - A
    elif subset == '11':
        elements = A & B
    ven.get_label_by_id(subset).set_text("\n".join(elements))
plt.show()
```

```python
from collections import Counter
import matplotlib.pyplot as plt


with open("temp.txt", 'r') as file:
    text = file.read()

words = text.split()
word_count = len(words)
print("The number of words in the file is: ",word_count)

frequency = dict(Counter(words))
frequent_word=""
word_frequency=0
for i in frequency:
    if frequency[i]>word_frequency:
        word_frequency=frequency[i]
        frequent_word=i

print("The most frequent word is ",frequent_word)


plt.bar(words, [ len(i) for i in words ])
plt.xlabel('Words')
plt.ylabel('Frequency')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats

numbers = [10, 20, 20, 30, 40, 40, 40, 50, 60, 70]

mean = np.mean(numbers)
median = np.median(numbers)
mode = list(stats.mode(numbers))[0]
std = np.std(numbers)

print("Mean: ", mean)
print("Median: ",median)
print("Mode: ",mode)
print("Standard Deviation: ",std)

plt.boxplot(numbers)
plt.ylabel('Value')
plt.show()
```

```python
import matplotlib.pyplot as plt

def normalize(lis):
    max_num = max(lis)
    min_num = min(lis)
    return [(x - min_num) / (max_num - min_num) for x in lis]

numbers = [1,10,20,30,40,50,60]

normalized_list = normalize(numbers)
print("Normalized numbers: "+str(normalized_list))

plt.plot(normalized_list)
plt.xlabel('No.of')
plt.ylabel('Normalized Value')
plt.show()
```
